### Methods and Plan

#### General Method and Plan
To predict attrition, I will use a logistic regression since attrition is a categorical variable. Attrition has a binary value, either "Yes'' or "No". Because of this the first step I will take will be to convert the Attrition attribute from a factor into a double, where 1 is representative of "Yes" and 0 is representative of "No". To perform the regression, I will use the glm function, predicting attrition using the input variable I described in the Exploratory Data Analysis part of this project. I will also be sure to specify the family of the regression as binomial, due to the nature of the Attrition variable being either 0 or 1 after mutation. For input data and testing purposes, I will be splitting the original data set into a 70% training and 30% testing set. I will use the testing set against the model created by the training data to test the fit of my model, and check if any of the variables can be removed from the model, reducing variance. Using tidy, I will transform these results into a more readable form, exponentiate them, and also compute a confidence interval for the true coefficients. Since the stakes in this data are relatively low (they are not life or death, merely if an employee decides to continue to work at a particular company), I will use a significance level of 0.10. 

#### Assumptions and Conditions
For a logistic model, we need to make some assumptions. The first of which is that our input variables can not have multicollinearity. We have already checked this condition in the Exploratory Data Analysis section in the heat map and this condition is met. Our data set also contains many attributes of an employee from both their work and personal life, which means it is unlikely that there will be any confounding or lurking variables as well. Lastly, we assume that our errors are independent, that is that the error in one prediction does not affect the error of another. We can also assume that these conditions are met as the data was randomly generated, and there was no effect on one observation to another in generation. This same fact also guarantees our final assumption, that each employee is independent of one another.

#### Limitations
One potential limitation of the methods I selected is not including interaction terms in the model. I do not believe that any of the variables will have a strong interaction with any of the others at this point and thus I have decided to leave the terms out of my model. Should evidence suggest that including interaction terms would lead to a better-fit model, I will test models with interaction terms against those without, but at this point, I am planning on leaving them out.




### Implementation of proposed model

First I will adjust the Attrition column, and make a bianry factor, and then split the data into a 70% trainng 30% testing set.

```r
set.seed(1)
attrition_data <- 
    employee_tidy %>% 
    mutate(Attrition = ifelse(Attrition == "Yes", 1, 0))

attrition_train <- 
    attrition_data %>% 
    slice_sample(prop = .7)

attrition_test <- 
    attrition_data %>% 
    anti_join(attrition_train, by = "EmployeeNumber") %>%
    select(-EmployeeNumber)

attrtion_train <- attrition_train %>%
    select(-EmployeeNumber)
```

To detemine the variable I should use for my model, I will be using lasso to help determine my input variables. This is not the method I orginally proposed, however we had not learned about lasso logistic regression at the time, and I feel it is better than brute forcing to find the optimal model.

First I will need to conver my explanatory and response varables to matricies, as this is what the glmnet package requires as arguments.

```r
set.seed(1)
matrix_X_train <- 
    model.matrix(Attrition~., attrition_train)[,-1]

matrix_Y_train <- 
    as.matrix(attrition_train$Attrition, ncol = 1)
    
matrix_X_test <- 
    model.matrix(Attrition~., attrition_test)[,-1]

matrix_Y_test <- 
    as.matrix(attrition_test$Attrition, ncol = 1)
```

We can now try to find the lambda for which the average AUC is maximized. We will use this value of lambda as opposed to the 1se lambda as we have already narrowed down our variables a significant amount, and there is no need to further reduce the amount of variables by being selective with lambda.

```r
set.seed(1)
attition_cv_lambda_lasso <- 
  cv.glmnet(
  x = matrix_X_train, y = matrix_Y_train,
  alpha = 1,
  family = "binomial",
  type.measure = "auc",
  nfolds = 5)

lambda <- round(attition_cv_lambda_lasso$lambda.min, 4)

attrition_lasso_1se_AUC <- glmnet(
  x = matrix_X_train, y = matrix_Y_train,
  alpha = 1,
  family = "binomial",
  lambda = lambda
)
```

It would seem from the lasso model above, that Age is the only varibale with a noticable effect on Attrition, so it will be the only variable I use when creating my model.

```r
set.seed(1)
Attrition_log_age <- 
    glm(formula = Attrition ~ Age, data = attrition_train, family = binomial)

tidy_version <- tidy(Attrition_log_age)
```

We will also plot the regression to see its effects visually

```r
set.seed(1)
age_attrition_plot <- 
    ggplot(data = attrition_train) +
    geom_point(aes(x = Age, y = Attrition)) +
    geom_smooth(aes(x = Age, y = Attrition),
        method = "glm",
        method.args = c(family = binomial), 
        se = FALSE, color = "red") +
    ylim(-0.5,1.5)  +
    ggtitle("Logisitic Regression, Attrition predicted by age") 
    # scale_x_continuous(breaks = seq(10, 90, 5))

age_attrition_plot
```

Now we will find the error rate. Since our model predicts everyone to not have attrition, we can do so without predict.

```r
set.seed(1)
people <- nrow(attrition_test)
attrition_no <- sum(attrition_test$Attrition == 0)
proportion = (attrition_no / people)
proportion
```
